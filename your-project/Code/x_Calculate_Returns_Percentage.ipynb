
The percentage of merchandise returned over a particular period can be calculated by simply dividing the number of items returned by the number that has been sold

```python
#Libraries
#Dataframe and arrays
import pandas as pd
import numpy as np
```

```python
df = pd.csv_read('../data/data_for_returns_analysis.csv')
```

```python
df['Rev'] = df['Quantity']*df['UnitPrice']
```

```python
df_returns = df.loc[df['Quantity'] < 0]
```

```python
df['Return'] = df.apply(lambda row: 1 if row.Quantity < 0 else 0, axis=1)
```

```python
df.head()
```

```python
df_returns.head()
```

```python
df['Sold'] = df.apply(lambda row: row.Quantity if row.Quantity > 0 else 0, axis=1)
```

```python
df['Returned'] = df.apply(lambda row: row.Quantity if row.Quantity < 0 else 0, axis=1)
```

```python
df.head()
```

```python
df_return_rate = df.groupby('StockCode').agg({'Sold': ['sum'],'Returned': ['sum']})
df_return_rate.reset_index(inplace=True)
```

```python
df_return_rate.columns = ['StockCode','Sold','Returned']
```

```python
df_return_rate.head()
```

```python
#df_return_rate['return_percent_unit'] = df.apply(lambda row: (row.Returned / row.Sold)*100, axis=1)

df_return_rate['return_percent_unit'] = df.apply(lambda row: ((row.Returned / row.Sold)*100) if row.Sold != 0 else 0, axis=1)


df_return_rate.head()
```

```python
df_return_rate['return_percent_unit'] = ( df_return_rate['Returned'] / df_return_rate['Sold'] ) * 100
```
